```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
movies=pd.read_csv('tmdb_movies.csv')
genres=pd.read_csv('tmdb_genres.csv')
genres.head()
```

```python
# Zwróć listę 10 najwyżej ocenianych filmów (vote_average), których liczba głosów (vote_count) jest większa od 3. kwartyla rozkładu liczby głosów.
```

```python
third_quantile=movies[movies['vote_count']>=(movies['vote_count'].quantile(.75))]
third_quantile.head()
```

```python
top_rated=third_quantile.sort_values(by='vote_average',ascending=False)
top_rated.iloc[:10,:]
```

```python
# Pogrupuj tabelę w taki sposób, aby otrzymać średni przychód (revenue) oraz średni budżet (budget) w danym roku dla filmów opublikowanych od 2010 (włącznie) do 2016 roku (włącznie). Następnie na tej podstawie stwórz wykres, w którym średnie przychody są wykresem kolumnowym, a średni budżet wykresem liniowym na tych samych osiach. Sformatuj odpowiednio oś X oraz oś Y. Dodaj tytuł wykresu, oraz legendę, która znajduje się w prawym górnym rogu płótna, lecz poza obszarem osi. Przykład wykresu widoczny poniżej:
```

```python
movies['release_date']=pd.to_datetime(movies['release_date'])
movies['year']=movies['release_date'].dt.year
movies.head()
```

```python
movies_2010_2016=movies[((movies['year']>=2010)&(movies['year']<=2016))]
movies_2010_2016.head()
```

```python
movies_grouped=movies_2010_2016.groupby('year').agg({'budget':'mean','revenue':'mean'})
movies_grouped.head()
```

```python
movies_grouped.reset_index(inplace=True)
```

```python
def million(x, pos):
        return 'USD {:2.0f}M'.format(x*1e-6)
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
x = movies_grouped['year']
y = movies_grouped['revenue']
z = movies_grouped['budget']
axes.plot(x,z,c='red',label='Budget')
axes.bar(x,y,label='Revenue')
formatter = plt.FuncFormatter(million)
axes.yaxis.set_major_formatter(formatter)
axes.legend(loc=(1.05,1.0))
axes.set_xlabel('Year')
axes.set_title("Average movie's budget and revenue by year")
plt.show()
```

```python
# Baza filmów zawiera kolumnę z id gatunku (genre_id). Na tej podstawie połącz ze sobą bazę filmów z bazą gatunków, tak aby w bazie filmów można było odczytać nazwę gatunku filmu.
```

```python
genres.columns=['genre_id','genre']
movies_genres=pd.merge(movies,genres,how='left',on='genre_id')
movies_genres.head()
```

```python
# Jaki gatunek filmu z bazy pojawia się w niej najczęściej? Ile filmów tego gatunku znajduje się w bazie?
```

```python
top_genre=movies_genres['genre'].value_counts()
top_genre.idxmax()
```

```python
# Filmy, którego gatunku trwają średnio najdłużej (runtime)?
```

```python
runtime=movies_genres.groupby('genre')['runtime'].mean().sort_values(ascending=False)
runtime.idxmax()
```

```python
# Stwórz histogram czasu trwania filmów z gatunku, który cechuje się największym średnim czasem trwania.
```

```python
movies_genres[movies_genres['genre']=='History']['runtime'].plot(kind='hist',bins=10)
```
